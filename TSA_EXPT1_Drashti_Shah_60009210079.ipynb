Detrending using Pandas

```python
from pandas import datetime
from matplotlib import pyplot
def parser(x):
  return datetime.strptime('190'+x, '%Y-%m')
series= pd.read_csv(r'shampoo.csv', index_col=0,parse_dates=True)
X= series.values
diff= list()
for i in range(1, len(X)):
  value= X[i] -X[i-1]
  diff.append(value)
pyplot.figure(figsize=(15,6))
pyplot.plot(diff)
pyplot.title('Detrending using Pandas Differencing', fontsize=16)
pyplot.xlabel('Year')
pyplot.ylabel('Sales exchange rate')
pyplot.show()
```

The variable 'X' stores the values from the 'series' and holds the exchange rate data. An empty list called 'diff' where we'll store the differences between consecutive exchange rate values. Initiates a for loop that iterates over the range of integers from 1 to the length of the 'X' array. value = X[i] - X[i - 1] will calculates the difference between the exchange rate value at the current index 'i' and the sales at the previous index 'i - 1'. This calculates the first-order difference in the time series data. diff.append(value) appends the calculated difference 'value' to the 'diff' list. Ploting the 'diff' list, which contains the differences between consecutive sales exchange rate over time. Assigning the title of the plot to 'Detrending using Pandas Differencing'. Detrending using Scipy Signal

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
df=pd.read_csv('/content/shampoo.csv', index_col=0,parse_dates=True)
detrended = signal.detrend(df.Sales.values)
plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title('Detrending using scipy signal',fontsize=16)
plt.show()
```

Importing the 'signal' module from the Scipy library, which provides various signal processing functions, including detrending. Assigning the title of the plot to 'Detrending using Scipy Signal'. Seting the label for the x-axis as 'Sales'and the y-axis as 'Frequency'.
